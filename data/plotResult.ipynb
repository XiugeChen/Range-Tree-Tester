```python
# read files
type1, type2 = [], []

fp = open (EXPERIMENT + ".log", 'r')
for line in fp:
    if "[ExperimentApp] Finish" in line:
        time = line.rstrip().split('=')[-1]
        
        if TYPE_1 in line:
            type1.append(int(time))
        elif TYPE_2 in line:
            type2.append(int(time))
        
df = pd.DataFrame({ 'x':X, TYPE_1:type1, TYPE_2:type2 })
```

```python
# plot results
plt.rcParams["figure.dpi"] = 128
plt.rcParams['figure.figsize'] = 10, 5
plt.plot('x', TYPE_1, data=df, linewidth=1)
plt.plot('x', TYPE_2, data=df, linewidth=1)
plt.title(TITLE)
plt.xlabel(X_LABLE)
plt.ylabel(Y_LABLE)
plt.legend(loc='upper left')
plt.show()
```
