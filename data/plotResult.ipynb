```python
# constants and variables
# experiment construct_time_data_length

EXPERIMENT = "construct_time_data_length"

TITLE = "Time v.s. Construction method"
Y_LABLE = "Actual construction time (in microseconds)"
X_LABLE = "Actual data length (* 10^3)"

TYPE_1 = "naive"
TYPE_2 = "smart"

X = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]

# experiment query_time_data_length
"""
EXPERIMENT = "query_time_data_length"

TITLE = "Query Time v.s. Data Length"
Y_LABLE = "Actual query time (in microseconds)"
X_LABLE = "Actual data length (* 10^3)"

TYPE_1 = "Original"
TYPE_2 = "Fractional Cascading"

X = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024]
"""
# experiment query_time_query_range
"""
EXPERIMENT = "query_time_query_range"

TITLE = "Query Time v.s. Query Range"
Y_LABLE = "Theoretical query time"
X_LABLE = "Theoretical query range"

TYPE_1 = "Original"
TYPE_2 = "Fractional Cascading"

X = [1, 2, 5, 10, 20]
"""
```

```python
# read files
type1, type2 = [], []

fp = open (EXPERIMENT + ".log", 'r')
for line in fp:
    if "[ExperimentApp] Finish" in line:
        time = line.rstrip().split('=')[-1]
        
        if TYPE_1 in line:
            type1.append(int(time))
        elif TYPE_2 in line:
            type2.append(int(time))
```

```python
# plot results
plt.rcParams["figure.dpi"] = 128
plt.rcParams['figure.figsize'] = 5, 5
plt.plot('x', TYPE_1, data=df, linewidth=1)
plt.plot('x', TYPE_2, data=df, linewidth=1)
plt.title(TITLE)
plt.xlabel(X_LABLE)
plt.ylabel(Y_LABLE)
plt.legend(loc='upper left')
plt.show()
```
