```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# constants and variables
EXPERIMENT = "construct_time_data_length"

TITLE = "Naive Construction v.s. Smart Construction"
Y_LABLE = "Time in milliseconds"
X_LABLE = "Length of data * 10^3"

TYPE_1 = "naive"
TYPE_2 = "smart"

X = ["2^1","2^2","2^3","2^4","2^5","2^6","2^7","2^8","2^9","2^10"]
```

```python
# read files
type1, type2 = [], []

fp = open (EXPERIMENT + ".log", 'r')
for line in fp:
    if "[ExperimentApp] Finish" in line:
        time = line.rstrip().split('=')[-1]
        
        if TYPE_1 in line:
            type1.append(int(time))
        elif TYPE_2 in line:
            type2.append(int(time))
        
df = pd.DataFrame({ 'x':X, TYPE_1:naive, TYPE_2:smart })
```

```python
# plot results
plt.rcParams["figure.dpi"] = 128
plt.rcParams['figure.figsize'] = 10, 5
plt.plot('x', TYPE_1, data=df, linewidth=1)
plt.plot('x', TYPE_2, data=df, linewidth=1)
plt.title(TITLE)
plt.xlabel(X_LABLE)
plt.ylabel(Y_LABLE)
plt.legend(loc='upper left')
plt.show()
```
